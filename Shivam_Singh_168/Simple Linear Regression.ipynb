```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

n = len(x)
sumx = 0
sumy = 0
sumxy = 0
sumx2 = 0
for i in range(n):
    sumx += x[i]
    sumy += y[i]
    sumxy += x[i] * y[i]
    sumx2 += x[i] * x[i]
m = (n * sumxy - sumx * sumy) / (n * sumx2 - sumx * sumx)
b = (sumy - m * sumx) / n

print("Slope (m):", m)
print("Intercept (b):", b)
xnew = 6
ypred = m * xnew + b

print("Predicted value for x =", xnew, "is:", ypred)
```
